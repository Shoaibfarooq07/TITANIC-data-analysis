```python
bins=[0 , 12 , 18 , 35 , 60 , 80]
labels=["child" , "teen" , "Young adult" ,"adult", "senior"]

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df
```

```python
survival_by_age_group=df.groupby(["AgeGroup"],observed=True)["Survived"].mean()
print(survival_by_age_group)
```

```python
age_group_survival = df.groupby('AgeGroup',observed=True)['Survived'].agg(['sum', 'count'])
print(age_group_survival)
```

```python
age_group_survivals = df.groupby('AgeGroup', observed=True)['Survived'].agg(['sum', 'count'])
print(age_group_survivals)
age_group_survivals.plot(kind='barh')


plt.xlabel('Survival Percentage')
plt.ylabel('Age Group')
plt.title('Survival Rate by Age Group')
plt.show()
```

```python
percentage=(age_group_survivals["sum"]/age_group_survivals["count"]) * 100
print(percentage)
percentage.plot(kind='barh',color=["red","blue","green","black","brown","purple"])


plt.xlabel('Survival Percentage')
plt.ylabel('Age Group')
plt.title('Survival Rate by Age Group')
plt.show()
```

```python
plt.barh(survival_by_age_group.index, survival_by_age_group.values, color=['darkblue','red','green','yellow','brown',])

plt.title('survival rate by age group')
plt.xlabel('survival rate')
plt.ylabel('age group')
plt.show
```

```python
survival_by_gender = df.groupby('Sex')['Survived'].agg(['sum', 'count'])
print(survival_by_gender)
```

```python
labels = survival_by_gender.index  
sizes = survival_by_gender['sum']  

# Plotting
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Survivors by Gender')
plt.axis('equal')  
plt.show()
```

```python
survival_by_gender=(survival_by_gender['sum'] / survival_by_gender['count']) * 100
print(survival_by_gender)
survival_by_gender.plot(kind='bar',color=["red","yellow"])

plt.ylabel('Survival Percentage')
plt.xlabel('Sex')
plt.title('Survival Rate by Gender')
plt.show()
```
